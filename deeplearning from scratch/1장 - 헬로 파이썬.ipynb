<h4> 2-3. 원소 접근 </h4>

인덱스를 배열로 지정해 한 번에 여러 원소에 접근 가능

```python
x = np.array([[51, 55], [14, 19], [0, 4]])
print (x)

for row in x :
    print (row)
```

```python
x = x.flatten() #x를 1차원 배열로 변환
print (x)

x[np.array([0, 2, 4])] #인덱스가 0, 2, 4인 원소 얻기
```

```python
x > 15
```

```python
x[x>15]
```

<h2> 3. matplotlib </h2>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)  #0에서 6까지 0.1 간격으로 생성
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.arange (0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = "sin")
plt.plot(x, y2, linestyle = "--", label = "cos")
plt.title ('sin&cos')
plt.show()
```
